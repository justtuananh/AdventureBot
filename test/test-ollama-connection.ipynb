```python
import requests

url = "https://asr.hpda.vn/stt"

audio_file_path = "/home/justtuananh/Downloads/speech.mp3"

# Ngôn ngữ
language = "Vietnamese"

# Mở tệp âm thanh và gửi dữ liệu
with open(audio_file_path, 'rb') as f:
    files = {'file': (audio_file_path, f)}

    # Dữ liệu form
    data = {'language': language}

    # Gửi yêu cầu POST
    response = requests.post(url, files=files, data=data)

    # Kiểm tra phản hồi
    if response.status_code == 200:
        print("Text from audio:", response.json()["text"])
    else:
        print("Request failed with status code:", response.status_code)
```

```python
import subprocess
import json

def generate_response(prompt, model="tinyllama", url="http://localhost:11434/api/generate"):
    """
    Generates a response based on the given prompt using the specified model.

    Args:
        prompt (str): The prompt to generate a response for.
        model (str): The name of the model to use for generation. Default is "tinyllama".
        url (str): The URL of the API endpoint for generating responses. Default is "http://localhost:11434/api/generate".

    Returns:
        str: The generated response.
    """
    try:
        # Tạo command để gửi yêu cầu CURL
        command = f"curl {url} -d '{{\"model\": \"{model}\", \"prompt\": \"{prompt}\"}}'"

        # Thực thi command và nhận output
        output = subprocess.check_output(command, shell=True)

        # Parse output từ dạng JSON
        response_data = json.loads(output)

        # Lấy phần response từ dữ liệu JSON
        generated_response = response_data.get("response", "Error: No response received")
        return generated_response
    except subprocess.CalledProcessError as e:
        return f"Error: {e}"

# Sử dụng hàm generate_response với prompt cụ thể
prompt = "Why is the sky blue?"
response = generate_response(prompt)
print("Generated Response:", response)
```
